```python
import yfinance as yf
```

```python
tesla = yf.Ticker('TSLA')
tesla_data = tesla.history(period='max')
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import requests
import pandas as pd
```

```python
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
response.raise_for_status()

tables = pd.read_html(response.content)

tesla_revenue = tables[1]
tesla_revenue.tail()
```

```python
gme = yf.Ticker('GME')
gme_data = gme.history(period='max')
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
response.raise_for_status()

tables = pd.read_html(response.content)

gme_revenue = tables[1]
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

def make_graph(stock_data, revenue_data, stock,  title):
  fig, ax1 = plt.subplots(figsize=(14, 6))
  ax1.plot(stock_data['Date'], stock_data['Close'], label='Stock Price')
  ax1.set_title(title)
  ax1.set_xlabel('Date')
  ax1.set_ylabel('Stock Price')
  ax1.legend()
  plt.show()


make_graph(tesla_data, tesla_revenue, 'Tesla', 'Tesla Stock Price')
```

```python
make_graph(gme_data, gme_revenue, 'GameStop', 'GameStop Stock Price')
```
